```python
tx_path = "02_Analisis_Financiero/datos/transacciones_ejemplo.csv"
if os.path.exists(tx_path):
    tx = pd.read_csv(tx_path)
    display(tx.head())
    # Ajusta nombres de columna según tu CSV
    if 'amount' in tx.columns and 'customer_id' in tx.columns:
        agg = tx.groupby('customer_id')['amount'].agg(['count','sum','mean']).sort_values('sum', ascending=False).head()
        display(agg)
    else:
        print("Revisa las columnas en transacciones:", tx.columns.tolist())
else:
    print("Archivo de transacciones no encontrado:", tx_path)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# ejemplo: histograma de amounts si existe
if 'tx' in globals() and 'amount' in tx.columns:
    plt.figure(figsize=(8,4))
    tx['amount'].dropna().hist(bins=30)
    plt.title("Distribución de amount")
    plt.xlabel("amount")
    plt.ylabel("frecuencia")
    plt.show()
else:
    print("No hay columna 'amount' para graficar.")
```
